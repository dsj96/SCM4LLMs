```python
import matplotlib.pyplot as plt
# fig, ax = plt.subplot()

scm_turbo = [65, 72]
scm_davinci003 = [67, 75]
scm_davinci003_wo_summary_gate = [64, 68]
scm_davinci003_wo_short_mem = [60, 65]
scm_davinci003_wo_long_mem = [64, 71]
plt.scatter(*scm_turbo, marker="o", color="purple", s=8**2, linewidths=2, label="$SCM_{turbo}$")
plt.scatter(*scm_davinci003, marker="*", color="r", s=12**2, linewidths=2, label="$SCM_{davinci003}$")
plt.scatter(*scm_davinci003_wo_summary_gate, marker="D", color="b", s=5**2, linewidths=2, label="$SCM_{davinci003}$ w/o memory controller")
plt.scatter(*scm_davinci003_wo_short_mem, marker=">", color="c", s=5**2, linewidths=2, label="$SCM_{davinci003}$ w/o flash memory")
plt.scatter(*scm_davinci003_wo_long_mem, marker=",", color="m", s=5**2, linewidths=2, label="$SCM_{davinci003}$ w/o activation memory")

plt.grid(True, alpha=0.2)

plt.legend()
plt.xlim(50, 75) # 控制坐标轴范围
plt.ylim(50, 80)
plt.xlabel("Win Rate of Coverage (%)", fontsize=14)
plt.ylabel("Win Rate of Coherence (%)", fontsize=14)
plt.title("Book Summarization", fontsize=14)
plt.savefig("./booksum.png", dpi=400)
```
